## 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def prime_or_not(num):
    '''
    this function is created to check if a number is prime or not
    arguments:
    num = int(numbers that will be checked)
    return:
    true if the number is prime otherwise false
    '''
    if num <= 1:
        return False #numbers less than or equal to 1 are not prime
    for i in range (2,int(num*0.5) + 1 ):
        if num%i == 0:
            return False #number is not prime 
    return True #number is prime 

for number in range(1,201):
    if prime_or_not(number):
        print(number," is a prime number")
```

## 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class FibonacciIterator:
    """
    An iterator class to generate the Fibonacci sequence up to a specified number of terms.
    """
    def __init__(self, num_terms):
        """
        Initialize the iterator with the number of terms.
        
        :param num_terms: Number of terms in the Fibonacci sequence to generate
        """
        self.num_terms = num_terms
        self.current = 0
        self.next = 1
        self.count = 0

    def __iter__(self):
        """Returns the iterator object."""
        return self

    def __next__(self):
        """Generates the next Fibonacci number."""
        if self.count >= self.num_terms:
            raise StopIteration  # Stop iteration when the specified number of terms is reached

        # Compute the current Fibonacci number
        fib_number = self.current
        # Update the values for the next iteration
        self.current, self.next = self.next, self.current + self.next
        self.count += 1
        return fib_number


# Example usage:
num_terms = 10
fib_iterator = FibonacciIterator(num_terms)

print(f"Fibonacci sequence up to {num_terms} terms:")
for fib_number in fib_iterator:
    print(fib_number, end=" ")
```

## 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_of_two(max_powers):
    '''
this generator function yields the power of 2 upto a given exponent 

arguments:
max_powers = it is a max exponent

yield:
will return the the power of 2 upto a given exponent 
'''
    for power in range (max_powers + 1):
        yield 2**power

max_powers = 10 
print(f"power of 2 up to {max_powers}")
for powers in power_of_two(max_powers):
    print(powers)


    
```

## 7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_line_by_line(file_path):
    '''
    this function read file line by line and yields each line as a string
    parameters:
    file_path is the path of the file which function will read line by line and will return each line as a string
    '''
    try:
        with open(file_path,'r') as file:
            for line in file:
                yield line.strip() 
    except FileNotFoundError:
        print(f"Error, File {file_path} was not found")
    except IOError:
        print(f"Error: an error occured while reading the file {file_path}")

file_path = "example.txt"
print("File content line by line")

for file in read_line_by_line(file_path):
    print(file)
```

## 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(input_string):
    vowels = 'aeiouAEIOU'
    no_vowels = filter(lambda char : char not in vowels, input_string)
    return "".join(no_vowels)

input_string = "This is the program that uses filter to remove all the vowels from a given string"
result = remove_vowels(input_string)
print("String before removing vowels:-",input_string)
print("String after removing vowels:-",result)
    
```
